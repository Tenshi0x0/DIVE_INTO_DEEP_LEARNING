## Q4

````python
import numpy as np

class DeterministicRandomData:
    def __init__(self, seed=42):
        self.seed = seed
        self.rng = np.random.default_rng(seed)
        self.fixed_state = self.rng.bit_generator.state

    def __call__(self, n=1000, d=2):
        self.rng.bit_generator.state = self.fixed_state
        X = self.rng.normal(size=(n, d))
        y = self.rng.normal(size=(n,))
        return X, y

gen = DeterministicRandomData(seed=114514)
X1, y1 = gen(5, 2)
X2, y2 = gen(5, 2)
np.array_equal(X1, X2), np.array_equal(y1, y2)
````
